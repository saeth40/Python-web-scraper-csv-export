```python
import urllib.request
import os
import glob
import requests
from requests.auth import AuthBase
from requests.auth import HTTPBasicAuth
from openpyxl import load_workbook, Workbook
from bs4 import BeautifulSoup
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
import urllib3
```

```python
# set up driver

driver = webdriver.Chrome("/Users/winit/Desktop/TDRI/chromedriver")


wait = WebDriverWait(driver,180)

driver.get('https://datawarehouse.dbd.go.th/')
time.sleep(30)
submit = driver.find_element_by_id('signinBtn')
submit.click()
time.sleep(15)

data = open('/Users/winit/Desktop/TDRI/dbd_new/46313.txt')
data = data.readlines()
data = [item.strip() for item in data]

indus_path = '/Users/winit/Desktop/TDRI/dbd_new/46313/'
if not os.path.exists(indus_path):
    os.makedirs(indus_path)

for item in data:
    
    url = item.strip().split('/')
    com = url[-1]
    b_type = url[-2]
    
    filename = indus_path + '/' + com + '_p_59.html'
    if os.path.exists(filename):
        continue
    
    
    
    info = 'https://datawarehouse.dbd.go.th/company/profile/' + b_type + '/' + com
    driver.get(info)
    element = wait.until(EC.element_to_be_clickable((By.XPATH, '/html/body/div/div[4]/div[2]/div[1]/div[2]/div[2]/div[1]')))
    html = driver.page_source
    filename = indus_path + '/' + com + '_info.html'
    with open(filename,'w',encoding='utf-8') as f:
        f.write(html)


    balance = 'https://datawarehouse.dbd.go.th/fin/balancesheet/' + b_type + '/' + com
    driver.get(balance)
    element = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="fixTable"]')))
    html = driver.page_source
    filename = indus_path + '/' + com + '_b.html'
    with open(filename,'w',encoding='utf-8') as f:
        f.write(html)

    balance_59 = 'https://datawarehouse.dbd.go.th/fin/balancesheet/' + b_type + '/' + com + '/2559'
    driver.get(balance_59)
    element = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="fixTable"]')))
    html = driver.page_source
    filename = indus_path + '/' + com + '_b_59.html'
    with open(filename,'w',encoding='utf-8') as f:
        f.write(html)

    profit = 'https://datawarehouse.dbd.go.th/fin/profitloss/' + b_type + '/' + com
    driver.get(profit)
    element = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="fixTable"]')))
    html = driver.page_source
    filename = indus_path + '/' + com + '_p.html'
    with open(filename,'w',encoding='utf-8') as f:
        f.write(html)

    profit_59 = 'https://datawarehouse.dbd.go.th/fin/profitloss/' + b_type + '/' + com + '/2559'
    driver.get(profit_59)
    element = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="fixTable"]')))
    html = driver.page_source
    filename = indus_path + '/' + com + '_p_59.html'
    with open(filename,'w',encoding='utf-8') as f:
        f.write(html)
    
    print(b_type,com)
    
driver.close()
```
